```python
# bias function

def bias_cal(guage, sat):
    bias= sum(guage - sat)/len(sat)
    return bias
```

```python
# correlation

def corl_cal(guage,sat):
    corr = np.corrcoef(guage, sat)
    return corr
```

```python
G_oda=GPCC_d['Oda']
G_Ada=GPCC_d['Ada']
G_Tema=GPCC_d['Tema']
G_Kumasi=GPCC_d['Kumasi']
```

```python
guage_d=[G_oda,G_Ada,G_Tema,G_Kumasi]

def splitting(data):
    m_oda=data['Oda']
    m_Ada=data['Ada']
    m_Tema=data['Tema']
    m_Kumasi=data['Kumasi']
    m_d=[m_oda,m_Ada,m_Tema,m_Kumasi]
    for gua,ml in guage_d and m_d:
        corrr=corl_cal(gua,ml)
    return corrr
```
