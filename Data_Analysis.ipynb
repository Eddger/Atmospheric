```python
# correlation calculation
```

```python
acrcorrp = stats.pearsonr(acrpes,acrgmt)
acrcorrpd = stats.pearsonr(acrddir,acrgt)
acrcorrcc = stats.pearsonr(acrccc,acrgtc)
acrcorrcd = stats.pearsonr(acrcdd,acrgtd)

adacorrp = stats.pearsonr(adapes,adagmt)
adacorrpd = stats.pearsonr(adaddir,adagt)
adacorrcc = stats.pearsonr(adaccc,adagtc)
adacorrcd = stats.pearsonr(adacdd,adagtd)

akcorrp = stats.pearsonr(akpes,akgmt)
akcorrpd = stats.pearsonr(akddir,akgt)
akcorrcc = stats.pearsonr(akccc,akgtc)
akcorrcd = stats.pearsonr(akcdd,akgtd)

spcorrp = stats.pearsonr(sppes,spgmt)
spcorrpd = stats.pearsonr(spddir,spgt)
spcorrcc = stats.pearsonr(spccc,spgtc)
spcorrcd = stats.pearsonr(spcdd,spgtd)

tmcorrp = stats.pearsonr(tmpes,tmgmt)
tmcorrpd = stats.pearsonr(tmddir,tmgt)
tmcorrcc = stats.pearsonr(tmccc,tmgtc)
tmcorrcd = stats.pearsonr(tmcdd,tmgtd)
```

```python
# mean calculation
```

```python
acrmnp = sum((acrpes-acrgmt)**2)/len(acrpes)
acrmnpd = sum((acrddir-acrgt)**2)/len(acrddir)
acrmcc = sum((acrccc-acrgtc)**2)/len(acrccc)
acrmcd = sum((acrcdd-acrgtd)**2)/len(acrcdd)

adamnp = sum((adapes-adagmt)**2)/len(adapes)
adamnpd = sum((adaddir-adagt)**2)/len(adaddir)
adamcc = sum((adaccc-adagtc)**2)/len(adaccc)
adamcd = sum((adacdd-adagtd)**2)/len(adacdd)

akmnp = sum((akpes-akgmt)**2)/len(akpes)
akmnpd = sum((akddir-akgt)**2)/len(akddir)
akmcc = sum((akccc-akgtc)**2)/len(akccc)
akmcd = sum((akcdd-akgtd)**2)/len(akcdd)

spmnp = sum((sppes-spgmt)**2)/len(sppes)
spmnpd = sum((spddir-spgt)**2)/len(spddir)
spmcc = sum((spccc-spgtc)**2)/len(spccc)
spmcd = sum((spcdd-spgtd)**2)/len(spcdd)

tmmnp = sum((tmpes-tmgmt)**2)/len(tmpes)
tmmnpd = sum((tmddir-tmgt)**2)/len(tmddir)
tmmcc = sum((tmccc-tmgtc)**2)/len(tmccc)
tmmcd = sum((tmcdd-tmgtd)**2)/len(tmcdd)
```

```python
# root mean squared error calculation
```

```python
acrpmse = np.sqrt(acrmnp)
acrdmse = np.sqrt(acrmnpd)
acrcmse = np.sqrt(acrmcc)
acrcdmse = np.sqrt(acrmcd)

adapmse = np.sqrt(adamnp)
adadmse = np.sqrt(adamnpd)
adacmse = np.sqrt(adamcc)
adacdmse = np.sqrt(adamcd)

akpmse = np.sqrt(akmnp)
akdmse = np.sqrt(akmnpd)
akcmse = np.sqrt(akmcc)
akcdmse = np.sqrt(akmcd)

sppmse = np.sqrt(spmnp)
spdmse = np.sqrt(spmnpd)
spcmse = np.sqrt(spmcc)
spcdmse = np.sqrt(spmcd)

tmpmse = np.sqrt(tmmnp)
tmdmse = np.sqrt(tmmnpd)
tmcmse = np.sqrt(tmmcc)
tmcdmse = np.sqrt(tmmcd)
```

```python
# bias calculation
```

```python
acrbiasp = sum(acrpes-acrgmt)/len(acrpes)
acrbiaspd = sum(acrddir-acrgt)/len(acrddir)
acrbiascc = sum(acrccc-acrgtc)/len(acrccc)
acrbiascd = sum(acrcdd-acrgtd)/len(acrcdd)

adabiasp = sum(adapes-adagmt)/len(adapes)
adabiaspd = sum(adaddir-adagt)/len(adaddir)
adabiascc = sum(adaccc-adagtc)/len(adaccc)
adabiascd = sum(adacdd-adagtd)/len(adacdd)

akbiasp = sum(akpes-akgmt)/len(akpes)
akbiaspd = sum(akddir-akgt)/len(akddir)
akbiascc = sum(akccc-akgtc)/len(akccc)
akbiascd = sum(akcdd-akgtd)/len(akcdd)

spbiasp = sum(sppes-spgmt)/len(sppes)
spbiaspd = sum(spddir-spgt)/len(spddir)
spbiascc = sum(spccc-spgtc)/len(spccc)
spbiascd = sum(spcdd-spgtd)/len(spcdd)

tmbiatm = sum(tmpes-tmgmt)/len(tmpes)
tmbiatmd = sum(tmddir-tmgt)/len(tmddir)
tmbiascc = sum(tmccc-tmgtc)/len(tmccc)
tmbiascd = sum(tmcdd-tmgtd)/len(tmcdd)
```

```python
# standard deviation calculation
```

```python
acrstdp = np.std(acrpes)
acrstdpd = np.std(acrddir)
acrstdcc = np.std(acrccc)
acrstdcd = np.std(acrcdd)

adastdp = np.std(adapes)
adastdpd = np.std(adaddir)
adastdcc = np.std(adaccc)
adastdcd = np.std(adacdd)

akstdp = np.std(akpes)
akstdpd = np.std(akddir)
akstdcc = np.std(akccc)
akstdcd = np.std(akcdd)

spstdp = np.std(sppes)
spstdpd = np.std(spddir)
spstdcc = np.std(spccc)
spstdcd = np.std(spcdd)

tmstdp = np.std(tmpes)
tmstdpd = np.std(tmddir)
tmstdcc = np.std(tmccc)
tmstdcd = np.std(tmcdd)
```
